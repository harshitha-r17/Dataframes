```python
import pandas as pd
df1=pd.read_csv("departments.csv")
df1
```

```python
df2=pd.read_csv("employees.csv")
df2
```

```python
innerjoin=pd.merge(df2,df1,on="Department_ID",how="inner")
innerjoin
```

```python
leftjoin=pd.merge(df2,df1,on="Department_ID",how="left")
leftjoin
```

```python
rightjoin=pd.merge(df1,df2,on="Department_ID",how="right")
rightjoin
```

```python
outerjoin=pd.merge(df2,df1,on="Department_ID",how="outer")
outerjoin
```

```python
totalsalary=df2.groupby("Department_ID")["Salary"].sum()
totalsalary
```

```python
avgsalary=df2.groupby("Department_ID")["Salary"].mean()
avgsalary
```
